```python
words_sentences = map_words_to_sentences(aoi_sentences_dict=aoi_sentences_dict, stimuli_sentences_dict=stimuli_sentences_dict)
```

```python
print("Words to Sentences Mapping:")
for word, sentence in words_sentences.items():
	print(f"  Word: {word}")
	print(f"    Sentence: {sentence}")
```

```python
# Add sentences to words_dict
for stimulus_key in words_dict:
	if stimulus_key in words_sentences:
		print(f"Adding sentences to words_dict for stimulus {stimulus_key}...")
		words = words_sentences[stimulus_key]
		for subject_id in words_dict[stimulus_key]:
			for word_idx in words_sentences[stimulus_key]:
				if word_idx in words_dict[stimulus_key][subject_id]:
					# If the word has fixations, add the sentence to the word entry
					words_dict[stimulus_key][subject_id][word_idx]['sentence'] = words[word_idx]['sentence']
				else:
					# If the word does not have fixations (not in words_dict), add it and set the sentence and the fixations empty
					words_dict[stimulus_key][subject_id][word_idx] = {
						'word': words[word_idx]['word'],
						'fixations': WordFixations(),
						'sentence': words[word_idx]['sentence']
					}
	else:
		print(f"Stimulus {stimulus_key} not found in words_sentences.")
```

```python
# Go through words_dict and make sure all words have a sentence
for stimulus_key in words_dict:
	print(f"Stimulus: {stimulus_key}")
	for subject_id in words_dict[stimulus_key]:
		for word_idx in words_dict[stimulus_key][subject_id]:
			if 'sentence' not in words_dict[stimulus_key][subject_id][word_idx]:
				words_dict[stimulus_key][subject_id][word_idx]['sentence'] = None
```

```python
# Print the updated words_dict with sentences
for stimulus_key in words_dict:
	print(f"Stimulus: {stimulus_key}")
	for subject_id in words_dict[stimulus_key]:
		for word_idx in words_dict[stimulus_key][subject_id]:
			print(f"  Subject ID: {subject_id}, Word Index: {word_idx}")
			word = words_dict[stimulus_key][subject_id][word_idx]['word']
			sentence = words_dict[stimulus_key][subject_id][word_idx]['sentence']
			print(f"Stimulus: {stimulus_key}, Subject ID: {subject_id}, Word Index: {word_idx}, Word: {word}, Sentence: {sentence}, Fixations: {words_dict[stimulus_key][subject_id][word_idx]['fixations']}")
```

```python
# Save the words_dict to a CSV file
output_file_path = 'word_sentence_fixations/words_dict_romanian_theo.csv'
os.makedirs(os.path.dirname(output_file_path), exist_ok=True)

with open(output_file_path, 'w', newline='', encoding='utf-8') as csvfile:
	fieldnames = ['stimulus', 'subject_id', 'word_index', 'word_index_in_sentence', 'word_id', 'word', 'sentence_index', 'sentence_id', 'sentence', 'fixations_list', 'fixations_TRT', 'surprisal']	
	writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
	writer.writeheader()

	for stimulus_key in words_dict:
		for subject_id in words_dict[stimulus_key]:
			for word_idx in words_dict[stimulus_key][subject_id]:
				word_idx = int(word_idx)
				word = words_dict[stimulus_key][subject_id][word_idx]['word']
				sentence = words_dict[stimulus_key][subject_id][word_idx]['sentence']
				fixations = words_dict[stimulus_key][subject_id][word_idx]['fixations']
				writer.writerow({
					'stimulus': stimulus_key,
					'subject_id': subject_id,
					'word_index': word_idx,
					'word': word,
					'sentence': sentence,
					'fixations_list': json.dumps(fixations.fixations),
					'fixations_TRT': json.dumps(fixations.TRT)
				})
	
# Sort the rows on stimulus, subject_id and word_idx
df = pd.read_csv(output_file_path)
df.sort_values(by=['stimulus', 'subject_id', 'word_index'], inplace=True)

# Add word_index_in_sentence to the CSV file
df['word_index_in_sentence'] = df.groupby(['stimulus', 'subject_id', 'sentence']).cumcount()

# Add sentence_index to the CSV file
# Go through the stimuli
for stimulus_key in words_dict:
	df.loc[df['stimulus'] == stimulus_key, 'sentence_index'] = pd.factorize(df.loc[df['stimulus'] == stimulus_key, 'sentence'])[0]

# Add word_id to the CSV file
df['word_id'] = df['stimulus'] + '_' + df['word_index'].astype(str)

# Add sentence_id to the CSV file
df['sentence_id'] = df['stimulus'] + '_' + df['sentence_index'].astype(int).astype(str)

df.to_csv(output_file_path, index=False)

print(f"words_dict saved to {output_file_path}")
```
